```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
num_col=[]
for i in df:
    if df[i].dtype == "int64":
        num_col.append(i)
df[num_col] = mms.fit_transform(df[num_col])
```

```python
df.head()
```

```python
df["gender"].replace(to_replace="Male",value=0,inplace=True)
df["gender"].replace(to_replace="Female",value=0,inplace=True)
df
```

```python
df["income"].replace(to_replace="<=50K",value=1,inplace=True)
df["income"].replace(to_replace=">50K",value=0,inplace=True)
df
```

```python
d_list = []
for i in df:
    if df[i].dtype == "object":
        d_list.append(i)
```

```python
workclass = pd.get_dummies(df["workclass"])
education = pd.get_dummies(df["education"])
marital_status = pd.get_dummies(df["marital-status"])
occupation = pd.get_dummies(df["occupation"])
relationship = pd.get_dummies(df["relationship"])
race = pd.get_dummies(df["race"])
native_country = pd.get_dummies(df["native-country"])
```

```python
df = pd.concat([df,workclass,education,marital_status,occupation,relationship,race,native_country],axis=1)
df
```
